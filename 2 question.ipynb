```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Segment'], df['Gross Sales'])
plt.title('Gross Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Gross Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Country'], df['Units Sold'])
plt.title('Units Sold by Country')
plt.xlabel('Country')
plt.ylabel('Units Sold')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot([df[df['Product'] == 'Carretera']['Sale Price'], df[df['Product'] == 'Montana']['Sale Price']], labels=['Carretera', 'Montana'])
plt.title('Sale Price by Product')
plt.ylabel('Sale Price')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Units Sold'], df['Manufacturing Price'])
plt.title('Units Sold vs. Manufacturing Price')
plt.xlabel('Units Sold')
plt.ylabel('Manufacturing Price')
plt.show()
```

```python
country_segments = df.pivot_table(index='Country', columns='Segment', values='Gross Sales', aggfunc='sum')
country_segments.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gross Sales by Country and Segment')
plt.xlabel('Country')
plt.ylabel('Gross Sales')
plt.legend(title='Segment')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Country', y='Sale Price')
plt.title('Sale Price Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Sale Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Discounts'], bins=15, edgecolor='black')
plt.title('Distribution of Discounts')
plt.xlabel('Discount Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
numerical_cols = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
